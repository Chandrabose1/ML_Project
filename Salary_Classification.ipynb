### Data Pre-Processing

In this step, the data is prepared for model. The data preparation in influenced by the inference from EDA.

1. The outliers are capped.
2. The Education, capital_gain, capital_loss features are removed.
3. Categorical aggregation is performed on column such as country, class 

```python
# function to cap the outlier
def outlier(col):
    q1=data[col].quantile(0.25)
    q3=data[col].quantile(0.75)
    iqr=q3-q1
    lower_range=q1-1.5*iqr
    upper_range=q3+1.5*iqr
    data[col]=np.where(data[col]<lower_range,lower_range,data[col])
    data[col]=np.where(data[col]>upper_range,upper_range,data[col])
```

```python
for i in numeric_col:
    outlier(i)
```

```python
for i in numeric_col:
    sns.boxplot(data[i])
    plt.show()
```

```python
# country
# ratio of people getting more than 50K in each country except US
ratio=(data[(data["country"].str.strip()!="United-States") & (data["income"].str.strip()==">50K")]["country"].value_counts()/data["country"].value_counts())
```

```python
temp=pd.DataFrame(ratio>=ratio.median()).reset_index()
temp.columns=["country","status"]
greater_than_median=temp[temp["status"]==True]["country"].values
greater_than_median
```

```python
temp=pd.DataFrame(ratio<ratio.median()).reset_index()
temp.columns=["country","status"]
lesser_than_median=temp[(temp["status"]==True)]["country"].values
lesser_than_median
```

```python
data["country"]=data["country"].replace(to_replace=greater_than_median,value="high_med_countries")
```

```python
data["country"]=data["country"].replace(to_replace=lesser_than_median,value="low_med_countries")
```

```python
data["country"]=data["country"].replace({" Outlying-US(Guam-USVI-etc)":"low_inc_countries"," Holand-Netherlands":"low_inc_countries"})
```
